### Задача №1.
Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

Одно из решений: можем найти искомое число возможностей через разность общего числа комбинаций из 52 карт по 4 и числа комбинаций из 52 карт по 4, где тузы отсутствуют. То есть:

$\normalsize\displaystyle N = C_{52}^4 - C_4^0\cdot C_{48}^4 = \frac{52!}{4!\cdot48!} - \frac{4!}{0!\cdot4!}\cdot\frac{48!}{4!\cdot44!} = \frac{49\cdot50\cdot51\cdot52}{4!} - \frac{45\cdot46\cdot47\cdot48}{4!} = 76145$

### Задача №2, 3.
Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Случайным образом студент вытягивает 3 вопроса. Какова вероятность, что все выбранные вопросы знакомы студенту? Какова вероятность что два из трёх вопросов знакомы студенту?

- Вероятность первого события $A_1$ (все выбранные вопросы знакомы студенту) найдем как отношение количества комбинаций из 50 знакомых вопросов по 3 к общему количеству комбинаций из 60 вопросов по 3. То есть:

$\normalsize\displaystyle A_1 = \frac{C_{50}^3}{C_{60}^3} = \frac{50!}{3!\cdot47!}\cdot\frac{3!\cdot57!}{60!} = \frac{48\cdot49\cdot50}{58\cdot59\cdot60} = \frac{980}{1711} \approx 0.573 = 57.3\%$

- Вероятность второго события $A_2$ (два из трех воросов знакомы студенту) найдем как отношение количества комбинаций из 50 знакомых вопросов по 2, умноженного на количество комбинаций из 10 незнакомых воросов по 1, к общему количеству комбинаций из 60 вопросов по 3. То есть:

$\normalsize\displaystyle A_2 = \frac{C_{50}^2\cdot C_{10}^1}{C_{60}^3} = \frac{50!}{2!\cdot48!}\cdot\frac{10!}{1!\cdot9!}\cdot\frac{3!\cdot57!}{60!} = \frac{49\cdot50\cdot10\cdot3}{58\cdot59\cdot60} = \frac{1225}{3422} \approx 0.358 = 35.8\%$

### Задача №4.
Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.   
Вопрос: какова вероятность, что вы действительно больны ей?

Обозначим вероятности:

$\normalsize P(D) = 0.001$ - вероятность заболеть,   
$\normalsize P\big(\bar{D}\big) = 0.999$ - вероятность не заболеть,   
$\normalsize P(T|D) = 0.99$ - вероятность положительного результата теста при условии заболевания человека,   
$\normalsize P\big(T|\bar{D}\big) = 0.01$ - вероятность положительного результата теста при условии, что человек не болен.

Тогда вероятность действительного заболевания человека при условии положительного результата теста найдем по формуле Байеса:

$\normalsize\displaystyle P(D|T) = \frac{P(T|D)\cdot P(D)}{P(T|D)\cdot P(D)+P\big(T|\bar{D}\big)\cdot P\big(\bar{D}\big)} = \frac{0.99\cdot0.001}{0.99\cdot0.001+0.01\cdot0.999} = \frac{11}{122} \approx 0.09 = 9\%$
